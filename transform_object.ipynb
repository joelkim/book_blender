# 오브젝트 모드 변형

- 오브젝트 모드에서 기본 변형은
  - Move
  - Rotate
  - Scale

## 변형 방법

- 메뉴
- 툴바
- 단축키

### 메뉴

- Object > Transform > Move/Rotate/Scale

### 툴바와 기즈모

- 툴바에서 Move, Rotate, Scale 툴을 선택하면 객체 선택시 객체에 기즈모가 표시됨
- 해당 기즈모를 드래그하여 변형

#### 객체 기즈모 고정

- Viewport Gizmo 툴에서 "Object Gizmos"의 Move/Rotate/Scale 체크하면 툴 선택과 관계없이 항상 트랜스폼 기즈모가 표시됨

#### 기즈모 사용시 조정 단축키

- control+드래그: 그리드 간격으로 조정
- shift+드래그: 세밀한 조정
- control+shift+드래그: 세밀한 그리드 간격으로 조정

## last action panel

- 변환을 하면 생기는 좌측 하단의 패널
- 방금한 변환을 파라미터 조정을 통해 바꿀 수 있음
- 정밀한 조정시 사용

## 단축키 

- 다음 단축키로도 변환
  - G + 조건 + 숫자 파라미터 + enter
  - R + 조건 + 숫자 파라미터 + enter
  - S + 조건 + 숫자 파라미터 + enter

- 조건
  - X
  - Y
  - Z
  - X > X
  - Y > Y
  - Z > Z
  - shift+X
  - shift+Y
  - shift+Z
  - shift+X > shift+X
  - shift+Y > shift+Y
  - shift+Z > shift+Z
  


## 사이드바

- 사이드바 > Item > Transform 
- 절대 위치로 변형 가능
