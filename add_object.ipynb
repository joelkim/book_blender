# 객체 추가

## 객체 추가

- 객체를 추가하는 기능
- 오브젝트 모드 뿐 아니라 에디트 모드에서도 사용 가능하지만 차이가 있음
- 오브젝트 모드에서 객체를 추가한 경우: 별개의 객체로 생성
- 에디트 모드에서 객체를 추가한 경우: 선택되어 있는 객체의 자식 객체로 생성

## 사용법

- 메뉴: 레이아웃 워크스페이스의 "Add" 메뉴 > "Mesh"

- 단축키: {{< kbd shift-a >}}

## 기본 생성 위치 및 크기

- 현재 3D Cursor 위치에 오브젝트 오리진이 오도록 생성됨
- Scene unit 기준으로 크기가 정해짐
