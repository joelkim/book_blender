# 단축키

## 모드 변환

| 키 | 기능 |
|-|----|
| {{< kbd Tab >}}                                       | 오브젝트 모드와 에디트 모드 변환 |



## UI 조작

| 키 | 기능 |
|-|----|
| {{< kbd t >}}                                         | 3D 뷰포트의 툴바 보이기/감추기 |
| {{< kbd n >}}                                         | 3D 뷰포트의 사이드바 보이기/감추기 |



## 선택

| 키 | 기능 |
|-|----|
| {{< kbd w >}}                                         | 선택도구를 돌려가면서 선택 |
| {{< kbd a >}}                                         | 모두 선택 |
| {{< kbd Alt >}} - {{< kbd a >}}                       | 선택 제외 |
| {{< kbd l >}}                                         | select linked |
| {{< kbd Shift >}} - {{< kbd l >}}                     | deselect linked |


## 파이 메뉴

| 키 | 기능 |
|-|----|
| {{< kbd Ctrl >}} - {{<kbd Tab >}}                     | pie menu - mode |
| {{< kbd Shift >}} - {{<kbd s >}}                     | pie menu - snap |



## 컨텍스트 메뉴

| 키 | 기능 |
|-|----|
| {{< kbd Shift >}} - {{< kbd a >}}                     | context menu - add |



## 편집

| 키 | 기능 |
|-|----|
| {{< kbd x >}}                                         | 객체 삭제 |
| {{< kbd F9 >}}                                        | 마지막 연산 조정 |



## 되돌리기

| 키 | 기능 |
|-|----|
| {{< kbd Ctrl >}} - {{<kbd z >}}                          | undo |
| {{< kbd Ctrl >}} - {{< kbd Shift >}} - {{<kbd z >}}      | redo |
