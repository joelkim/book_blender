# 단축키

## Mode Change

| 키 | 기능 |
|-|----|
| {{< kbd Tab >}}                                       | object mode $\leftrightarrow$ edit mode 



## UI

| 키 | 기능 |
|-|----|
| {{< kbd t >}}                                         | 3D Viewport toolbar show/hide |
| {{< kbd n >}}                                         | 3D Viewport sidebar show/hide |



## Select

| 키 | 기능 |
|-|----|
| {{< kbd a >}}                                         | select all |
| {{< kbd Alt >}} - {{< kbd a >}}                       | deselect all |
| {{< kbd l >}}                                         | select linked |
| {{< kbd Shift >}} - {{< kbd l >}}                     | deselect linked |


## Pie Menu

| 키 | 기능 |
|-|----|
| {{< kbd Ctrl >}} - {{<kbd Tab >}}                     | pie menu - mode |
| {{< kbd Shift >}} - {{<kbd s >}}                     | pie menu - snap |



## Context Menu

| 키 | 기능 |
|-|----|
| {{< kbd Shift >}} - {{< kbd a >}}                     | context menu - add |



## Edit

| 키 | 기능 |
|-|----|
| {{< kbd x >}}                                         | delete selected |



## Undo/Redo

| 키 | 기능 |
|-|----|
| {{< kbd Ctrl >}} - {{<kbd z >}}                          | undo |
| {{< kbd Ctrl >}} - {{< kbd Shift >}} - {{<kbd z >}}      | redo |
