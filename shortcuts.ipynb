# 단축키

| 키 | 기능 |
|-|----|
| {{< kbd Fn-Ctrl-Up >}}   | prev workspace |
| {{< kbd Fn-Ctrl-Down >}} | next workspace |
| {{< kbd Tab >}}          | object mode $\leftrightarrow$ edit mode |
| {{< kbd Ctrl-Tab >}}     | pie menu |
| {{< kbd t >}}            | 3D Viewport toolbar show/hide |
| {{< kbd n >}}            | 3D Viewport sidebar show/hide |
| {{< kbd shift-space >}}  | 3D Viewport toolbar context menu |
| {{< kbd shift-a >}}      | add |
| {{< kbd a >}}            | select all |
| {{< kbd option-a >}}     | deselect all |
| {{< kbd l >}}            | select linked |
| {{< kbd shift-l >}}      | deselect linked |
