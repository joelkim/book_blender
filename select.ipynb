# 선택

## 오브젝트 모드에서 선택

- 기본 선택 방법: 마우스 클릭
  - 선택된 오브젝트 경계가 오렌지색이 됨

- 추가선택: Shift + 마우스 클릭
  - 새로 선택된 오브젝트의 경계가 오렌지색
  - 이미 선택되어 있던 오브젝트의 경계는 붉은색

## 툴바 선택 메뉴

- Tweak
- Select Box
- Select Circle
- Select Lasso

### 셀렉트 모드

- 트윅을 제외한 나머지 셀렉트 방식에서는 셀렉트 모드 툴이 나타남

- Set: 디폴트 모드. 선택할 때마다 선택된 객체들이 바뀜
- Extend: 선택할 때마다 추가. 기존 선택 + 추가 선택
- Subtract: 기존 선택에서 제외. 기존 선택 - 추가 선택
- Difference: 기존 선택 + 추가 선택 집합에서 중복 선택된 것은 제외
- Intersect: 기존 선택 + 추가 선택 집합에서 중복 선택된 것만 선택

### 트윅 선택

- 마우스 클릭으로 선택
- Shift+클릭: 추가

### 박스 선택

- 마우스 "클릭-드래그"해서 생기는 사각형의 경계에 걸리는 오브젝트를 선택
- Shift+선택: 추가
- Contrl+선택: 제거

### 서클 선택

- 오브젝트 오리진이 원 안에 들어오는 경우에만 선택 
- Shitf+선택: 추가
- Contrl+선택: 제거

- 오브젝트 오리진 표시 방법
  - "Viewport Overlays > Objects > Origins(All)" 선택

- 선택 원 크기 조정

## 에디트 모드에서 선택

- 에디트 모드에서는 vertex, edge, face 구분

## 엑스레이

- 엑스레이 모드에서는 뒷쪽의 vertext, edge 등도 선택 가능
- face는 face origin이 가장 가까운 face가 선택됨
