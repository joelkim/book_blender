# 돌출

- 에디트 모드에서만 사용 가능
- 오브젝트를 구성하는 점, 선, 면을 바깥으로 나오게 돌출시키는 새로운 점, 선, 면을 생성함
- 기존 점, 선, 면을 이동시키는 것이 아님
- 점을 돌출시키면
  - 새로운 점과 그 점을 잇는 선을 생성하지만 실제로 잘 사용되지 않음
- 선을 돌출시키면
  - 새로운 면과 그 면을 구성하는 점, 선을 생성함
- 면을 돌출시키면
  - 새로운 볼륨과 그 볼륨을 구성하는 점, 선, 면을 생성함

## 사용법

1. 에디트 모드 진입
2. 돌출시킬 점, 선, 면 선택
3. E 키를 눌러 돌출 모드 진입
4. 마우스 등으로 세부 지정
