# 여러개의 오브젝트

## Joinning

- 제한사항
  - 같은 유형의 오브젝트만 join 가능

### Boolean

- join 연산은 메쉬를 결합하지 않음
- 메쉬를 결합하려면 Boolean 연산을 해야 함

## Duplicate Linking

- duplicated link 된 객체는 메시 데이터를 공유함
- 하나를 edit 모드에서 수정하면 다른 객체도 공유
- 따라서 transform 되어도 항상 전반적인 모양이 같음
- object mode에서 복수를 선택해도 edit mode로 가면 하나만 편집 가능

### 이미 존재하는 객체를 duplicated link하는 법

- 객체를 멀티 선택
- 마지막에 선택된 active 오브젝트가 소스가 된다.
- "Object" > "Link/Transfer Data" > "Link Object Data"

### duplicated link되어있는 객체를 분리하는 법 

- "Object" > "Relations" > "Make Single User" 

## Parenting

- 부모를 transform 하면 자식에게 동시 적용
- 자식을 transform 하면 부모에게 적용되지는 않음

### Parenting 방법

- 객체를 멀티 선택
- 마지막에 선택된 active 오브젝트가 parent가 된다.
- "Object" > "Parent" > "Object"

### parent 해제하는 법

- 자식 선택
- "Object" > "Parent" > "Clear Parent"
