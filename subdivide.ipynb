# 분할

## 객체 분할(Subdivide)

- 오브젝트 모드에서는 사용 불가, 에디트 모드에서만 사용 가능
- 에디트 모드에서 점/선/면 선택후 우클릭 -> Subdivide


## 분할 대상

- 선을 기본 대상으로 함
- 점을 선택하는 경우, 이어진 점으로 구성된 선을 대상으로 함
- 면을 선택하는 경우, 면을 구성하는 선을 대상으로 함

## 사각형 면의 선이 선택된 경우

### 사각형 면을 구성하는 선 중 하나만 선택된 경우

- 점 선택 모드에서 서로 이어진(linked) 2개 이상의 점을 선택하거나 선 선택 모드에서 하나의 선을 선택한 경우
- 선을 분할함
  

### 사각형 면을 구성하는 선 중 연결된 2개의 선이 선택된 경우

- quad corner type 옵션에 따라 달라짐
  - inner vert
  - path
  - straight cut
  - fan

#### straight cut

- 각 선을 분할하고 각각 연결하여 삼각형과 사각형을 생성

#### path

- straight cut과 비슷하지만 마지막 점도 연결하여 하나의 사각형을 더 생성

#### fan

- 맞은편의 점으로부터 생성된 각 점을 연결하여 부채꼴 생성

#### inner vert

- stragit cut으로 만들어진 삼각형과 사각형 내부에 하나의 점을 더 만들고 맞은편 점과 연결

### 사각형 면을 구성하는 선 중 마주보는 2개의 선이 선택된 경우

- 무조건 두 선분을 분할하고 마주보는 점끼리 연결하여 선을 면을 분할

### 사각형 면을 구성하는 선 중 연결된 3개의 선이 선택된 경우

### 사각형 면을 구성하는 선 중 연결된 4개의 선이 선택된 경우

- 무조건 4개의 선을 분할하여 격자모양의 면을 생성
